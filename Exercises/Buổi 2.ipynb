```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
x= np.linspace(0,1,201)
y = np.sin((2*np.pi*x)**2)
plt.plot(x,y,'blue')
plt.show()
```

```python
x=10*np.random.rand(200,1)
y = (0.2+0.8*x)*np.sin(2*np.pi*x)+np.random.randn(200,1)
plt.scatter(x,y,c="FireBrick")
plt.show()
```

```python
x=10*np.random.rand(200,1)
y = (0.2+0.8*x)*np.sin(2*np.pi*x)+np.random.randn(200,1)
plt.hist(y,bins=20)
plt.show()
```
