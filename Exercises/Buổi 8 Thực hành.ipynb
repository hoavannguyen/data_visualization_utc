```python
from bokeh.io import output_file,show
from bokeh.plotting import figure
from bokeh.models import ColumnDataSource
from bokeh.sampledata.iris import flowers as dataframe
```

```python
import numpy as np
import pandas as pd
```

```python
source = ColumnDataSource(dataframe)
```

```python
dataframe.head()
```

```python
source1 = ColumnDataSource(dataframe)
p1 = figure()
p1.circle('sepal_length','sepal_width',source= source1) 
#show(p1)
```

```python
source2 = ColumnDataSource(dataframe)
p2 = figure()
p2.circle('petal_length','petal_width',source= source2)
#show(p2)
```

```python
source3 = ColumnDataSource(dataframe)
p3 = figure()
p3.circle('sepal_length','petal_width',source= source3)
#show(p3)
```

```python
from bokeh.layouts import row,column
from bokeh.plotting import figure, output_file,show,output_notebook
output_notebook()
layout1 = row(p1,p2,p3)
output_file('row.html')
show(layout1)
```

```python
layout2 = column(p1,p2,p3)
output_file('column.html')
show(layout2)
```

```python
layout3 = row(column(p1,p2),p3)
output_file('nested.html')
show(layout3)
```

```python
from bokeh.layouts import gridplot
layout4= gridplot([[None,p1],[p2,p3]],toolbar_location=None)
output_file('nested.html')
show(layout4)
```

```python
from bokeh.models.widgets import Tabs,Panel
first = Panel(child=row(p1,p2),title='First Tab')
second = Panel(child=column(p2,p3),title='Tab 2')
tabs = Tabs(tabs=[first,second])
output_file=('tabs.html')
show(tabs)
```

```python
p1.x_range=p2.x_range=p3.x_range
layout5 = row(p1,p2,p3)
show(layout5)
```

***Anotation and Guides***

```python
from bokeh.models import CategoricalColorMapper
mapper = CategoricalColorMapper(
        factors= ['setosa','virginica','versicolor'],
        palette=['red','green','blue'])
p = figure()
p.circle('sepal_length','sepal_width',
        size=10, source=source, color={'field':'species','transform':mapper},
        legend='species')
#p.legend.location='top_left'
```

```python
show(p)
```

```python
from bokeh.models import HoverTool
hover= HoverTool(
        tooltips=[
            ('name','@species'),
            ('x','@sepal_length'),
            ('y','@sepal_width')
        ])
p = figure(tools=[hover,'pan','wheel_zoom','reset'])
p.circle('sepal_length','sepal_width',
        size=10, source=source, color={'field':'species','transform':mapper},
        legend='species')
p.legend.location='bottom_right'
show(p)
```
