```python
# Câu 1
m = len(flowers.columns)
n = len(flowers.index)
print(f"Dataset có {n} quan sát, {m} thuộc tính")
```

```python
# Câu 2
print('Định lượng:')
for i in range(len(flowers.columns)):
    if flowers.dtypes[i] == 'float64':
        print('\t',flowers.columns[i],'  float64')
print('Định tính:')
for i in range(len(flowers.columns)):
    if flowers.dtypes[i] == 'object':
        print('\t',flowers.columns[i],'  object')
```

```python
# Câu 3 
import matplotlib.pyplot as plt

plt.hist(flowers['sepal_length'], bins = 20)
# Khai báo tiêu đề cho trục x
plt.title('Histogram Hòa by Nguyễn Văn Hòa')
plt.xlabel('species')
# Khai báo tiêu đề cho trục y
plt.ylabel('iris sepal length (cm)')
plt.show()
```

```python
# Câu 4
plt.scatter(flowers['sepal_length'],flowers['petal_length'])
plt.title('Scatter plot Sepal Length and Petal Length Iris by Nguyễn Hòa')
plt.show()
```

```python
# Câu 5
pd.plotting.scatter_matrix(
    flowers[flowers.columns[0:4]],
    alpha=0.75,
    figsize = (10,10))
plt.title("Scatter Matrix Iris by Nguyễn Hòa")
plt.show()
```

```python
# Câu 6
from bokeh.io import output_file,show
from bokeh.plotting import figure
from bokeh.models import ColumnDataSource
from ipywidgets import interact, interact_manual
from bokeh.plotting import figure, output_file,show,output_notebook
output_notebook()
```

```python
features = flowers.columns[0:4]
@interact(feature1=features,feature2=features)
def uif(feature1,feature2):
    source = ColumnDataSource(flowers)
    p1 = figure( plot_width=500, plot_height=500)
    p1.circle(feature1,feature2,source= source) 
    layout= p1
    show(layout)
```
